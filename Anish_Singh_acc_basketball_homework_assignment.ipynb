```python
import pandas as pd

acc_df = pd.read_csv('acc_players-2324F.csv', skiprows = 1)
acc_df.head(5)
```

```python
total_points = acc_df['PTS'].sum()
print("The total points scored by all the players in the acc was", total_points, "points.")
```

```python
most_min_df = acc_df[['Player', 'MP']].sort_values(by = "MP", ascending = False).head(1)
print(most_min_df.iloc[0]['Player'], "achieved the most minutes played by player in the acc, with a total of", most_min_df.iloc[0]['MP'], "minutes.")
```

```python
top_rebounders_df = acc_df[['Player', 'TRB']].sort_values(by = "TRB", ascending = False).head(5)
print("The top five rebounders in the acc are:", top_rebounders)
```

```python
filter_mins_df = acc_df[acc_df['MP'] > 500]
filter_mins_df
```

```python
highest_ast_over_500_df = filter_mins_df[['Player', 'AST']].sort_values(by = "AST", ascending = False).head(1)
print(highest_ast_over_500_df.iloc[0]['Player'], "achieved the most rebounds out of all players over 500 minutes, with a total of", highest_ast_over_500_df.iloc[0]['AST'], "rebounds.")
```

```python
top_three_ast_df = acc_df[['Player', 'AST']].sort_values(by = "AST", ascending = False).head(3)
print("The top three assist leaders in the acc are:", top_three_ast_df)
```

```python
top_three_blk_df = acc_df[['Player', 'BLK']].sort_values(by = "BLK", ascending = False).head(3)
print("The top three assist leaders in the acc are:", top_three_blk_df)
```

```python
points_by_school = acc_df.groupby('School')['PTS'].sum()
print("The total points scored by each school in the acc are:")
points_by_school.head(100)
```

```python
assists_by_school = acc_df.groupby('School')['AST'].sum()
print("The total assists achieved by each school in the acc are:")
assists_by_school.head(100)
```

```python
top_three_schools_by_points_scored = acc_df.groupby('School')['PTS'].sum().sort_values(ascending = False).head(3)
print("The top three schools by points scored are:")
top_three_schools_by_points_scored
```

```python
top_scorers_df = acc_df[['Player', 'PTS']].sort_values(by = "PTS", ascending = False).head(5)
import matplotlib.pyplot as plt
import seaborn as sns

sns.barplot(x = 'Player', y = 'PTS', data = top_scorers_df)
plt.title('Top Scorers in the ACC')
plt.xlabel('Player')
plt.ylabel('Points Scored')
plt.show()
```

```python
import numpy as np
sns.scatterplot(x = 'FG%', y = 'WS', data = acc_df)
plt.title('Field Goal Percentage vs Win Shares')
plt.xlabel('Field Goal Percentage')
plt.ylabel('Win Shares')
plt.show()
```

```python
correlation_coefficient = np.corrcoef(acc_df['WS'], acc_df['PTS'])[0, 1]
print("The correlation coefficient is", correlation_coefficient, "which indicates there is a strong positive correlation between win shares and field goal percentage.")
```
